`Data Features:`

**total_bill**: a numeric vector, the bill amount (dollars)

**smoker**: a factor with levels No Yes, whether the party included smokers


**day**: a factor with levels Friday Saturday Sunday Thursday, day of the week


**time**: a factor with levels Day Night, rough time of day


**size**: a numeric vector, number of people in a table/Party

### Retreiving the Dataset. Checking info and shape. Displaying first 5 rows.

```python
#Importing the data set on tips for exploratory analysis into a pandas dataframe, "tips_df"
tips_df = pd.read_csv("tips.csv")

## display the first 5 rows
print(tips_df.head(5))



```

```python
##checking dimesnion of the data
print(tips_df.shape) # tips_df dataframe has 244 rows and 8 columns

## more descriptive analysis of the attributes of the data frame

print(tips_df.describe()) 
''' Each attribute has 244 entries.
The average bill is 19.79 dollars.
On average, a tip is about 3.00 dollars
For any given meal, a table has about 2 people.
'''
# The 5 point number summary is clear from the description of the dataframe for the total_bill, tip and size attributes; the quantitaive data.
```

### Understanding the distribution of 'tip' and 'total_bill' using different plots


####  a. Using distribution plots

```python
## Performing univariate analysis for total_bill, tips and size.
## Using the distplot first to get a in depth visual of the distribution. 

# distplot without kde for total_bill
sns.distplot(tips_df["total_bill"],bins=40,kde=False)

# from the plot we can tell that the bills typically lie between 10 and 20 dollars. The count for bills paid for a meal priced beyond 20 dollars actiely reduces.
```

```python
# distplot with kde for tips
sns.distplot(tips_df["tip"],kde=True)

# from this we can conclude that most tips lie between 1 and 4 dollars

```

```python
# distplot for size
sns.distplot(tips_df["size"],kde=True)

# the highest frequency for tips lies at 2 dollars.

```

#### b. Using boxplots

```python
## Using boxplot to understand the distribution.

## using seaborn for more refined visuals
 # 1. Checking the relation between tipps and smoking
plt.figure(figsize=(12, 9))
sns.boxplot(data=tips_df, x = 'smoker',y='tip');

""" Data on non smokers has more outliers than that for smokers.
Boxplot for smokers is left skewed. This shows a higher tip amount when people are not smoking compared with those who do."""
```

```python
 # 2. Checking the realtion between total_bill and gender. For this we will determine which gender eats more expensive meals and vice versa.

plt.figure(figsize=(12, 9))
sns.boxplot(data=tips_df, x = 'sex',y='total_bill');

""" both have outliers. 
However, the median bill amount paid for males is greater than that for females. This would indicate that males have more expensive or comprehensive meals than females."""

```

### Relationship between 'total_bill' and 'tip'


#### a. Jointplot

```python
## relationship between total_bill(explanatory variable) and tip(response variable)

# default, scatterplot
sns.jointplot(x="total_bill",y="tip",data=tips_df)
```

```python
### using hexagonal plots by adding the kind argument
sns.jointplot(x="total_bill",y="tip",data=tips_df,kind="hex")

# from this we can conclude that tips are greatly concentrated below 3 dollars for 10-40 dollar meals. This can also be concluded using a kde plot.
```

```python
### histplot with a regression line for total bill and tip
sns.jointplot(x="total_bill",y="tip",data=tips_df,kind="reg")

# from this we can tell that tips tends to increase with an increase in the bill payable. Expensive meals= greater tips
```

#### b. Pairplots

```python
## simple pairplot for the dataframe tips_df
sns.pairplot(tips_df)

# difficult to make inferences from this plot on the categorical attributes such as sex, smoker, day and time. We can enforce the hue argument and different colur codes for this
```

```python
## hue argument for pairplots of the tips dataframe
#1. Sex
sns.pairplot(tips_df,hue="sex")

""" males are better tippers than females.
More males eat at the restaurant and have meals that are more costly compared to females"""
```

```python
# 2. Smoker
sns.pairplot(tips_df,hue="smoker",palette="magma")

""" Smokers tip more than non smokers
Smokers have more expensive bills than non smokers."""
```

```python
### 3. Day
sns.pairplot(tips_df,hue="day",palette="inferno")

""" Friday has the lowest sales for the restaurant with the least total bills paid.
Saturday records the highest amounts in tips
On saturdays, most tables have more than 2 people. This is in line with the fact that Saturday has the highest amount in total_bill paid"""
```

```python
## 4. Time
sns.pairplot(tips_df,hue="time",palette="plasma")

# we can conclude that
""" The highest traffic into the restaurant is over dinner
Tips are aslo more generous over dinner compared to lucnh
Customers also tend to have more expensive or comprehensive meals over dinner compared to lunch"""
```

#### c. Lmplot

```python
## a linear model plot for the relationship between total_bill and tip
sns.lmplot(data=tips_df,x="total_bill",y="tip")

# tips have a postive corelation with the total bill payable for a meal.
```

```python
## we could also want to know whether the number of people sharing a table affects their tipping and if so how it affects the tip
sns.lmplot(data=tips_df,x="size",y="tip")

# there is no clear linear relation between size and tips. While a realtion exists, I'm not quite sure which one.
```

### Distribution of tip with different gender. 


```python
## boxplot

sns. boxplot(data=tips_df,x ="sex",y = "tip")

""" men are not only more likely to tip but are also better tippers. This is because from the box plot, male barplot has a greater median than the female barplot. Male boxplot shows that the 
data on male customers is right skewed from the difference in the length of whiskers."""
```

### Relationship between 'total_bill', 'tip' and 'sex'


```python
## relationship between total_bill and tip
sns.lmplot(data=tips_df,x="total_bill",y="tip" )

## tips tend to be greater for higher bills.
```

```python
## sex is a categorical attribute. We therfore need to encode it if we are to use it.

# adding a new column to the data frame "sex_encoded" where a 0=female and 1=male
tips_df['sex_encoded'] = tips_df['sex'].map({'Female': 0, 'Male': 1})

# display first five entries
tips_df.head(5)
```

```python
## we can now use the sex_encoded column for manipulation

# relationship between total bill and sex

sns.lmplot(data=tips_df,y="total_bill",x="sex_encoded" )

# men spend larger sums on food compared to women at the restaurant
# this is not a linear relationship.

```

```python
### relationship between tips and sex
sns.lmplot(data=tips_df,y="tip",x="sex_encoded" )

# the relationship between tips ad sex is not linear.
```

### Investigating Which day brings in higher bill amount?


```python
## stripplot

sns. stripplot( data=tips_df,y="day",x="total_bill",jitter=0.05,alpha=0.3,size=7,linewidth=1,palette="bright");

# saturday brings in higher bill amounts compared to other days of the week.
```

```python
## swarmplot

sns.swarmplot(data=tips_df,x="day",y="total_bill",palette="deep",alpha=1,size=7,linewidth=1);

# saturday brings in higher bill amounts for the business.

```

```python
## boxplot
sns.boxplot(data=tips_df,x='day',y='total_bill',palette='Set2');

# what is the essece of the semicolon at the end of the code, the code seems to run with or without it?
```

### Gender Distribution of visits.


```python
# count of males and females visting each day

sns.countplot(data=tips_df,x="sex",hue="day",palette="Set1")

# We can clealry see that male traffic into the restaurant is more than female.
```

### Time of day that records highest sales 



```python
## time of the day is a categorical attribute and we have to encode it to use a linear model plot
# adding a new column to the data frame "visit_time" where a 0=Dinner and 1=Lunch
tips_df['visit_time'] = tips_df['time'].map({'Dinner': 0, 'Lunch': 1})

# display first 5 rows
print(tips_df.to_string())

```

```python
## linearmodel plot to determine which day the restaurant gets higher bills for males

sns.lmplot(data=tips_df[tips_df['sex'] == 'Male'],x="visit_time",y="total_bill")

# for male customers,the restaurant collects a higher amount in meals over dinner compares to lunch
```

```python
## linearmodel plot to determine which day the restaurant gets higher bills for females

sns.lmplot(data=tips_df[tips_df['sex'] == 'Female'],x="visit_time",y="total_bill")

# for female customers,the restaurant collects a higher amount in meals over dinner compares to lunch
```
